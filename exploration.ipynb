```python
import sys
import unicodedata

tbl = dict.fromkeys(i for i in range(sys.maxunicode)
                    if unicodedata.category(chr(i)).startswith('P') or
                    unicodedata.category(chr(i)).startswith('S') or
                    unicodedata.category(chr(i)).startswith('C'))

def remove_punctuation(text):
    return text.translate(tbl)
```

```python
import re, pprint, string
text = " ".join(list(df['ADDL_INFORMATION'].dropna().values))
text = remove_punctuation(text)
tokens = nltk.word_tokenize(text.translate(dict.fromkeys(string.punctuation)))

# of the 15 people who said additional info about water damage, common trends are roof leaking/damage
# carpets affected by water damage
# len(finance_ppl['WATER_DESC'].dropna())
# for entry in finance_ppl['WATER_DESC'].dropna().iteritems():
#     print (entry)

# finance_vehicle_dmg = finance_ppl.loc[finance_ppl[vehicle_dmg] == 'Yes']
# len(finance_vehicle_dmg)
```

```python
no_stopwords = [token.lower() for token in tokens if token.lower() not in stopwords.words('english')]
```
