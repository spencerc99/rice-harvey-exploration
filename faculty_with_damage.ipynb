```python
%matplotlib inline
import numpy
import pandas as pd
import matplotlib

import sys
# reload(sys)  
# sys.setdefaultencoding('utf8')

# name, netid, department, flags on housing or vehicle damage for all faculty

timestamp = '20170902'

survey_data = pd.DataFrame.from_csv('NA_ALL_' + timestamp + '.csv', encoding = "ISO-8859-1")

survey_data = survey_data[(survey_data['VEHICLE_DAMAGE'] == 'Yes') | (survey_data['WATER_DAMAGE'] == 'Yes')]
survey_data['HomeDepartment'] = survey_data['STUDENT_COLL_DEPT'].fillna(survey_data['DEPARTMENT'])
survey_data['RICE_EMAIL_ADDRESS'] = survey_data['RICE_EMAIL_ADDRESS'].fillna(survey_data['NetID'] + '@rice.edu')
columns_of_interest = ['NetID', 'HomeDepartment', 'RICE_EMAIL_ADDRESS', 'FULL_NAME', 'WATER_DAMAGE', 'WATER_DESC', 'VEHICLE_DAMAGE', 'VEHICLE_DESC', 'HOME_PHONE', 'EMPLOYEES_PR_HOME_PHONE', 'PHONE_AVAIL', 'PHONE_INFO']
survey_data[columns_of_interest].head()
```

```python
writer = pd.ExcelWriter('ReportedVehicleDamage.xlsx')
survey_data[columns_of_interest].drop_duplicates().to_excel(writer, index = False)
writer.save()
```
